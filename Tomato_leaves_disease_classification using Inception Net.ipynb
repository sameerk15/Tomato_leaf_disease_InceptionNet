```python
from tensorflow.keras.layers import Input, Lambda, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.inception_v3 import InceptionV3
#from keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.models import Sequential
import numpy as np
from glob import glob
```

```python
IMAGE_SIZE = [224, 224]
```

```python
train_path='G:/archive (1)/New Plant Diseases Dataset(Augmented)/train'
valid_path='G:/archive (1)/New Plant Diseases Dataset(Augmented)/valid'
```

```python
inception = InceptionV3(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
inception.layers
```

```python
for layer in inception.layers:
    layer.trainable = False
```

```python
folders=glob('G:/archive (1)/New Plant Diseases Dataset(Augmented)/train/*')
```

```python
folders
```

```python
inception.output
```

```python
x = Flatten()(inception.output)
```

```python
prediction = Dense(len(folders),activation='softmax')(x)
```

```python
model = Model(inputs=inception.input, outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,shear_range = 0.2, zoom_range=0.2, horizontal_flip=True)
```

```python
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
training_set=train_datagen.flow_from_directory('G:/archive (1)/New Plant Diseases Dataset(Augmented)/train',target_size=(224,224),batch_size=32,class_mode='categorical')
```

```python
test_set=test_datagen.flow_from_directory('G:/archive (1)/New Plant Diseases Dataset(Augmented)/valid',target_size=(224,224),batch_size=32,class_mode='categorical')
```

```python
r=model.fit(training_set, validation_data=test_set,epochs=20,steps_per_epoch=len(training_set),validation_steps=len(test_set))
```
